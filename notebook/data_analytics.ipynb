```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#summarize the data dataframe
print("Summary of the data frame:")
df.info()
```

```python
#statistical inference of the dataframe
print("mathematical view of the dataframe:")
df.describe()
```

```python
#what is the distribution of customers in  dataset
#how are customers distributed by gender and region
plt.figure(figsize=(10, 7))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title("Age Distribution of Customers")  # Corrected this line
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Gender'].value_counts()
```

```python
gender_count = df['Gender'].value_counts()
plt.figure(figsize=(10, 5))
gender_count.plot(kind='bar', color=['pink', 'blue'])
plt.title("Distribution of Customers Based on Gender")  # Corrected this line
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#distribution based on region
df['Region'].value_counts()
```

```python
region_distribution=df['Region'].value_counts()
plt.figure(figsize=(10,7))
region_distribution.plot(kind='bar',color=['orange'])
plt.title(" customer distribution of Regions")
plt.xlabel('Region')
plt.ylabel('count')
plt.show()
```

```python
region_avg_cltv = df.groupby('Region')['CLTV'].mean()
plt.figure(figsize=(8, 5))
region_avg_cltv.sort_values().plot(kind='bar', color='purple')
plt.title("Average CLTV by Region")  
plt.xlabel('Region')
plt.ylabel('Average CLTV')
plt.show()
```

```python
highest_cltv=df.groupby('Region')['CLTV'].max()
highest_cltv
```

```python
plt.figure(figsize=(8, 5))
highest_cltv.sort_values().plot(kind='bar', color='red')
plt.title("Highest  CLTV by Region")  
plt.xlabel('Region')
plt.ylabel('Average CLTV')
plt.show()
```
